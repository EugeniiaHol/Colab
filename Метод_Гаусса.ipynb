```python
def print_system(a, b, step=""):
    print(f"\nСтан системи {step}")
    for i in range(len(a)):
        row = "  ".join(f"{a[i][j]:>7.3f}" for j in range(len(a[i])))
        print(f"{row} | {b[i]:.3f}")
    print()

def gauss_method(a, b):
    n = len(b)

    # Прямий хід (елімінація)
    for i in range(n):
        # Знаходимо головний елемент у колонці
        max_el = abs(a[i][i])
        max_row = i
        for k in range(i + 1, n):
            if abs(a[k][i]) > max_el:
                max_el = abs(a[k][i])
                max_row = k

        # Обмін рядків
        a[i], a[max_row] = a[max_row], a[i]
        b[i], b[max_row] = b[max_row], b[i]

        print_system(a, b, f"після обміну рядків {i} та {max_row}")

        # Занулення нижніх елементів
        for k in range(i + 1, n):
            c = -a[k][i] / a[i][i]
            for j in range(i, n):
                if i == j:
                    a[k][j] = 0
                else:
                    a[k][j] += c * a[i][j]
            b[k] += c * b[i]

        print_system(a, b, f"після занулення в стовпці {i}")

    # Зворотний хід
    x = [0 for _ in range(n)]
    for i in range(n - 1, -1, -1):
        x[i] = b[i] / a[i][i]
        for k in range(i - 1, -1, -1):
            b[k] -= a[k][i] * x[i]

        print(f"Знайдено x[{i}] = {x[i]:.3f}")

    return x

# Вихідна система
A = [
    [2, -1, 3, 2],
    [3, 3, 3, 2],
    [3, -1, -1, 2],
    [3, -1, 3, -1]
]
B = [11, 16, 4, 6]

result = gauss_method(A, B)
print("\nРозв'язок системи:")
for i, val in enumerate(result):
    print(f"x[{i}] = {val:.3f}")
```
